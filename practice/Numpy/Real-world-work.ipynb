# Practice with real world data

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Advance operation with Bussiness example

```python
# [Resturant_id,2020,2021,2022,2023,2024]
sales_data=np.array([
                     [1,150000,180000,250000],# shop-1
                     [2,120000,140000,190000],# shop-2
                     [3,200000,260000,300000],# shop-3
                     [4,180000,210000,270000],# shop-4
                     [5,160000,185000,230000],# shop-5
                     ])
print("=== sales analysis ===")
print("\n sales data shape:",sales_data.shape)
print("\nSample data for 1st 3 result:\n",sales_data[:3])
print("\nSample data for 1st 3 result:\n",sales_data[:,1:])
```

### Total sales per year

```python
print(np.sum(sales_data, axis=0))
yearly_total=np.sum(sales_data[:,1:],axis=0)
print(yearly_total)
```

### min sales data

```python
min_sales=np.min(sales_data, axis=0)
print(min_sales)
```

### maximum sales per year

```python
max_sales=np.max(sales_data,axis=0)
print("Maximum sales:",max_sales)
```

### average sales per restautant

```python
avg_sales=np.mean(sales_data[:, 1:],axis=1)
print(avg_sales)
```

cummlative sales

```python
cumsum= np.cumsum(sales_data[:, 1:],axis=1)
print(cumsum)
plt.figure(figsize=(8,6))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Average cumlative sales:")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
